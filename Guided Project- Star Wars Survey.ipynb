```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
star_wars = pd.read_csv("star_wars.csv", encoding="ISO 8859-1")
star_wars.head()
```

```python
star_wars.columns
```

```python
pd.set_option("display.max_columns", None)
star_wars.head(20)
```

```python
star_wars = star_wars.loc[pd.notnull(star_wars.RespondentID),:]
star_wars
```

```python
star_wars.shape
```

```python
Yes_No = {"Yes":True, "No":False}
cols = ["Have you seen any of the 6 films in the Star Wars franchise?"
       ,"Do you consider yourself to be a fan of the Star Wars film franchise?"]
for c in cols:
    star_wars[c] = star_wars[c].map(Yes_No)

star_wars.head()
```

```python
cols_6= star_wars.columns[3:9]
cols_6
```

```python
m = star_wars.ix[1, range(3,9)]
m
```

```python
for i, c in enumerate(cols_6):
    T_F = {m[i]:True, np.nan:False}
    star_wars[c] = star_wars[c].map(T_F)
star_wars
```

```python
for i, c in enumerate(cols_6):
    star_wars = star_wars.rename(columns={c:"seen_" + str(i+1)})
star_wars
```

```python
star_wars[star_wars.columns[9:15]].dtypes
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars[star_wars.columns[9:15]].dtypes
```

```python
cols_names = star_wars.columns[9:15]
cols_names
```

```python
for i, c in enumerate(cols_names):
    star_wars = star_wars.rename(columns={c:"ranking_" + str(i+1)})
star_wars    
```

```python
avg_ranking = star_wars[star_wars.columns[9:15]].mean()
avg_ranking
```

```python
%matplotlib inline
fig, ax = plt.subplots()
ax.bar(range(1,7), avg_ranking, align= "center", width = 0.6,)
ax.tick_params(top="off", right="off")
for i, spine in ax.spines.items():
    spine.set_visible(False)
ax.set_ylim((0,4.5))
```

The plot above shows these respondents prefer the series 4, 5, 6 which are the original movies to the series 1, 2, 3 which are released laterly.

```python
total_seen = star_wars.loc[:, star_wars.columns[3:9]].sum()
ax = total_seen.plot.bar()
ax.set_ylabel("Number of people saw the movie")
```

It shows the original movies are seen by more respondents. And, it confirms what we saw in the ranking plot that people prefer oringinal movies more than the newer one.

```python
star_wars.columns
```

```python
fans = star_wars.loc[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True, :]
no_fans = star_wars.loc[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False, :]
```

```python
fans[fans.columns[9:15]].mean().plot.bar()
plt.show()

no_fans[no_fans.columns[9:15]].mean().plot.bar()
plt.show()
```

In the survey of all star wars fans, it's obvious that they love the original movie series more than the newer series.

However, the preferance of non star wars fans are not so clear even though they all consider series five as the most favorite one.

```python
fans[fans.columns[3:9]].sum().plot.bar()
plt.show()

no_fans[no_fans.columns[3:9]].sum().plot.bar()
plt.show()
```

These two plots show that series five and six are seen a lot by both fans of star wars and non fans.

However, series one to four are seen more by fans group compared to the non fans group.
